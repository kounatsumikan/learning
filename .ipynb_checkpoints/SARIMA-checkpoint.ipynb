# SARIMAモデル

ARIMAモデル(自己回帰和分移動平均モデル)
AR(自己回帰モデル)
MA(移動平均モデル)
I(和分過程)
を組み合わせたもの。
ある時系列データに対して,「d回差分」を取った定常的なデータは，「p個以前の過去の値」と「q個以前のノイズの値」によって現在の値を記述することができるモデル
このモデルは、ARIMA(p,d,q)と記述される。

目的変数説明変数の作成

```python
X_dataset1 = apple2["2008-03-27":"2018-03-27"]
```

```python
#
diff = X_dataset1.diff()
diff = diff.dropna()
```

```python
diff.plot()
```

```python
# Graph data
fig, axes = plt.subplots(2, 1, figsize=(15,16))

fig = sm.graphics.tsa.plot_acf(diff, lags=40, ax=axes[0])
fig = sm.graphics.tsa.plot_pacf(diff, lags=40, ax=axes[1])
```

```python
# 差分系列への自動ARMA推定関数の実行
resDiff = sm.tsa.arma_order_select_ic(diff, ic='aic', trend='nc')
resDiff
```

ARIMA(1, 1, 1)

## 総当たりで、AICが最小となるSARIMAの次数を探す
※メモリエラーで実行不可でした

## 季節変動性の時数を決め打ちで実行

```python
# SARIMAモデルを「決め打ち」で推定する
sarima = sm.tsa.SARIMAX(diff, order=(3,1,2), seasonal_order=(1,1,1,60)).fit()
print(sarima.summary())
```

```python
# 残差のチェック
residSARIMA = sarima.resid
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(residSARIMA.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(residSARIMA, lags=40, ax=ax2)
```

```python
# 予測
pred = sarima.predict("2016-03-27","2018-03-27")

# 実データと予測結果の図示
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(diff["2008-03-27":"2018-03-27"])
ax.plot(pred["2016-03-27":"2018-03-27"], "r:")# 
```

# ランダムフォレスト

```python
# クラス読み込み
from sklearn.ensemble import RandomForestClassifier
```

```python
#使用するデータ
dataset = pd.DataFrame()

dataset["apple_C"] = apple["Close"]["2005-1-28":]
dataset["apple_O"] = apple["Open"]["2005-1-28":]
dataset["apple_H"] = apple["High"]["2005-1-28":]
dataset["apple_L"] = apple["Low"]["2005-1-28":]

dataset["intel_C"] = intel["Close"]["2005-1-28":]
dataset["intel_O"] = intel["Open"]["2005-1-28":]
dataset["intel_H"] = intel["High"]["2005-1-28":]
dataset["intel_L"] = intel["Low"]["2005-1-28":]

dataset["wti_C"] = wti["Close"]["2005-1-28":]
dataset["wti_O"] = wti["Open"]["2005-1-28":]
dataset["wti_H"] = wti["High"]["2005-1-28":]
dataset["wti_L"] = wti["Low"]["2005-1-28":]

dataset["ms_C"] = ms["Close"]["2005-1-28":]
dataset["ms_C"] = ms["Open"]["2005-1-28":]
dataset["ms_C"] = ms["High"]["2005-1-28":]
dataset["ms_C"] = ms["Low"]["2005-1-28":]
```

```python
dataset
```

```python
# 欠損値の有無を確認
dataset.isnull().any() # 欠損値があるか確認
```

## モデリング用のデータの準備

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# --- 訓練データとテストデータに分割 ------

# 特徴量と目的変数に分ける
X_dataset = dataset
y_dataset = apple["difference"]["2005-1-28":].astype("str")

# 特徴量名を取り出しておく
feature_x_dataset = X_dataset.columns
feature_y_dataset = y_dataset.name

# 学習データと評価データに分ける
X_dataset_train, X_dataset_test, y_dataset_train, y_dataset_test = train_test_split(X_dataset, y_dataset, test_size=0.2, random_state=0) 
print("shape of X and y for training: ", X_dataset_train.shape, y_dataset_train.shape)
print("shape of X and y for testing: ", X_dataset_test.shape, y_dataset_test.shape)
```

## 学習

```python
clf_rf = RandomForestClassifier(random_state=0)
clf_rf.fit(X_dataset_train, y_dataset_train)
```

## 予測

```python
y_rf_pred = clf_rf.predict(X_dataset_test)
```

## 精度確認

```python
cm_rf = confusion_matrix(y_dataset_test, y_rf_pred)
acc_rf = accuracy_score(y_dataset_test, y_rf_pred)

print("confusion matrix: \n", cm_rf)
print("accuracy: \n", acc_rf)
```
